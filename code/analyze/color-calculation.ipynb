```python
import numpy as np
import pandas as pd
import PIL
import matplotlib.pyplot as plt
import statistics
from matplotlib.cbook import get_sample_data
import os
```

```python
cwd = os.getcwd()
```

```python
dataset = pd.read_csv("data/semmem-vs-memcat.csv")
```

```python
# Z to A
dataset_desc = dataset.sort_values(by=["rank_semmem"], ascending=False)
```

```python
# A to Z
dataset_asc = dataset.sort_values(by=["rank_semmem"], ascending=True)
```

# All Histogram


```python
image_for_proccess = 50
```

```python
def get_avg(dataset):
    return statistics.mean(dataset)
```

## ASC


```python
images = []
image_size = []
counter = 0
for index, row in dataset_asc.iterrows():

    if counter <= image_for_proccess:
        counter += 1
    else:
        break

    im = PIL.Image.open(cwd + "/images/" + row["SemMem"])
    image_size.append(im.size)
    images.append(im.load())
```

```python
red = []
green = []
blue = []

for i in range(0, 256):
    red.append(0)
    green.append(0)
    blue.append(0)
```

```python
counter = 0
for im in images:
    im_size = image_size[counter]
    counter += 1
    for i in range(0, im_size[0]):
        for j in range(0, im_size[1]):
            image_value = im[i, j]
            red[image_value[0]] += 1
            green[image_value[1]] += 1
            blue[image_value[2]] += 1
```

```python
round(get_avg(red))
```

```python
round(get_avg(blue))
```

```python
round(get_avg(green))
```

```python
plt.plot(red, "r")
plt.savefig("red-50-forgettable.jpg")
plt.show()
```

```python
plt.plot(blue, "b")
plt.savefig("blue-50-forgettable.jpg")
plt.show()
```

```python
plt.plot(green, "g")
plt.savefig("green-50-forgettable.jpg")
plt.show()
```

```python
red_forgettable = red
green_forgettable = green
blue_forgettable = blue
```
