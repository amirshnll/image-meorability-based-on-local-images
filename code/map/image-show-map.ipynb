```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
data = pd.read_csv("data/data.csv")
data_sorted = data.sort_values(by="rank")
```

```python
rank_number = []
files_count = 8

for key, item in data_sorted[len(data_sorted) - files_count :].iterrows():
    rank_number.append({"img": item["image"], "rank": item["rank"]})

path = "map/"
images = []

for item in rank_number:
    image = item
    img_path = path + image["img"]

    temp_rank = img_path.replace(path, "")

    images.append({"img": img_path, "rank": image["rank"]})

plt.figure()
_, fig = plt.subplots(2, 4, figsize=(20, 10))
fig = fig.flatten()
i = 0


for f in fig:
    f.imshow(mpimg.imread(images[i]["img"]))
    f.axis("off")
    f.title.set_text(images[i]["rank"])

    i += 1

plt.subplots_adjust(bottom=0.3)
plt.savefig("image/semmem-top.jpg")
```
