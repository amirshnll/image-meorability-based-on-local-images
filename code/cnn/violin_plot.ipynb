```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set_style("whitegrid")
```

```python
semmem = pd.read_csv("data/semmem.csv")
memcat = pd.read_csv("data/memcat.csv")
sub_memcat = pd.read_csv("data/sub_memcat.csv")
```

```python
resnet50 = pd.read_csv("csv/predictions_resnet50.csv")
resnet101 = pd.read_csv("csv/predictions_resnet101.csv")
vrnet = pd.read_csv("csv/predictions_vresnet50scratch.csv")
```

## Real SemMem

```python
semmem_ax = sns.violinplot(saturation=1, data=semmem, x="category", y="rank_semmem", palette="muted" )
plt.savefig('plot/semmem_ax.png')
```

## MemCat

```python
memcat_ax = sns.violinplot(saturation=1, data=memcat, x="category", y="rank", palette="muted" )
plt.savefig('plot/memcat_ax.png')
```

## Sub MemCat

```python
sub_memcat_ax = sns.violinplot(saturation=1, data=sub_memcat, x="category", y="rank_memcat", palette="muted" )
plt.savefig('plot/sub_memcat_ax.png')
```

## ResNet 50

```python
resnet50_ax = sns.violinplot(saturation=1, data=resnet50, x="category", y="prediction", palette="muted" )
plt.savefig('plot/resnet50_ax.png')
```

## ResNet 101

```python
resnet101_ax = sns.violinplot(saturation=1, data=resnet101, x="category", y="prediction", palette="muted" )
plt.savefig('plot/resnet101_ax.png')
```

## VRNet

```python
vrnet_ax = sns.violinplot(saturation=1, data=vrnet, x="category", y="prediction", palette="muted" )
plt.savefig('plot/vrnet_ax.png')
```

## get Average

```python
semmem_avg = semmem["rank_semmem"].mean()
memcat_avg = memcat["rank"].mean()
sub_memcat_avg = semmem["rank_memcat"].mean()
resnet50_avg = resnet50["prediction"].mean()
resnet101_avg = resnet101["prediction"].mean()
vrnet_avg = vrnet["prediction"].mean()
```

```python
print("semmem_avg: ", semmem_avg)
print("memcat_avg: ", memcat_avg)
print("sub_memcat_avg: ", sub_memcat_avg)
print("resnet50_avg: ", resnet50_avg)
print("resnet101_avg: ", resnet101_avg)
print("vrnet_avg: ", vrnet_avg)
```

## Compare

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(15,5))
ax1.set_title("real")
ax2.set_title("prediction")

semmem_ax = sns.violinplot(data=semmem, x="category", y="rank_semmem", ax=ax1, color="dodgerblue")
resnet50_ax = sns.violinplot(data=resnet50, x="category", y="prediction", ax=ax2, color="lightcoral")

ax1.axhline(y = semmem_avg, color = 'blue', linestyle = '-')
ax2.axhline(y = resnet50_avg, color = 'red', linestyle = '-')

plt.savefig('plot/semmem-vs-resnet50.png')

plt.show()
```
